```python
# Dataset info
# school - student's school (binary: 'SS' - 수시 or 'JS' - 정시)
# sex - student's sex (binary: 'F' - female or 'M' - male)
# age - student's age (numeric: from 22 to 27)
# address - student's home address type (binary: 'U' - urban or 'R' - rural)
# famsize - family size (binary: 'LE3' - less or equal to 3 or 'GT3' - greater than 3)
# Mjob - mother's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')
# Fjob - father's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other’)
# reason - reason to choose this school (nominal: close to 'home', school 'reputation', 'course' preference or 'other')
# traveltime - home to school travel time (numeric: 1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour)
# studytime - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours) 
# failures - number of past class failures(F) (numeric: n if 1<=n<3, else 4)
# internet - Internet access at home (binary: yes or no)
# romantic - with a romantic relationship (binary: yes or no)
# famrel - quality of family relationships (numeric: from 1 - very bad to 5 - excellent)
# freetime - free time after school (numeric: from 1 - very low to 5 - very high)
# goout - going out with friends (numeric: from 1 - very low to 5 - very high)
# Dalc - workday alcohol consumption (numeric: from 1 - very low to 5 - very high)
# Walc - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)
# health - current health status (numeric: from 1 - very bad to 5 - very good)
# absences - number of school absences (numeric: from 0 to 93)
# These grades are related with the course subject, Math or Portuguese:
# G1 - first period grade (numeric: from D0 to A+)
# G2 - second period grade (numeric: from D0 to A+)
# G3 - final grade (numeric: from D0 to A+)
# label - major (computer = 0 / ai = 1)

# 따로 읽어온 두 개의 데이터를 합치기
# comptuer전공 학생의 정보와 ai전공 학생의 정보를 합치기
data = pd.concat([computer, ai])

# 필요없는 열 제거
# 23개의 피처와 1개의 label을 갖는 데이터가 만들어짐
data.drop(['Unnamed: 0'], axis = 1, inplace = True)
data.head()
```

```python
data = pd.concat([computer, ai])
data = data.iloc[:,1:]
data.head()
```

```python
# 데이터 전처리 : 데이터 타입이 object인 경우 int형 또는 float형으로 데이터를 변환
# label의 경우에도 Computer는 0, AI의 경우는 1을 부여

# school (SS = 0, JS = 1)
data['school'] = data['school'].map({'SS':0, 'JS':1}).astype(int)

# sex (M = 0, F = 1)
data['sex'] = data['sex'].map({'M':0, 'F':1}).astype(int)

# address (U = 0, R = 1)
data['address'] = data['address'].map({'U':0, 'R':1}).astype(int)

# famsize (LE3 = 0, GT3 = 1)
data['famsize'] = data['famsize'].map({'LE3':0, 'GT3':1}).astype(int)

# Mjob, Fjob (teacher = 0, health = 1, services = 2, at_home = 3, other = 4)
jobs = {"teacher":0, "health":1, "services":2, "at_home":3, "other":4}
data['Mjob'] = data['Mjob'].map(jobs)
data['Fjob'] = data['Fjob'].map(jobs)

# reason (course = 0, home = 1, reputation = 2, other = 3)
data['reason'] = data['reason'].map({'course':0, 'home':1,'reputation':2, 'other':3}).astype(int)

# internet - yes의 값만 존재함으로 피처를 제거
data.drop(['internet'], axis = 1, inplace = True)

# romantic (no = 0, yes = 1)
data['romantic'] = data['romantic'].map({'no':0, 'yes':1}).astype(int)

#grade(A+:8, A0:7, B+:6, B0:5, C+:4, C0:3, D+:2, D0:1)
grade_to_num = {"A+":8, "A0":7, "B+":6, "B0":5, "C+":4, "C0":3, "D+":2, "D0":1}
data['G1'] = data['G1'].map(grade_to_num)
data['G2'] = data['G2'].map(grade_to_num)
data['G3'] = data['G3'].map(grade_to_num)

# grade_to_num = {"A+":4.5, "A0":4.0, "B+":3.5, "B0":3.0, "C+":2.5, "C0":2.0, "D+":1.5, "D0":1.0}
# data['G1'] = data['G1'].map(grade_to_num)
# data['G2'] = data['G2'].map(grade_to_num)
# data['G3'] = data['G3'].map(grade_to_num)

# label (Computer = 0, AI = 1)
data['label'] = data['label'].map({'Computer':0, 'AI':1}).astype(int)
```

### Support Vecotr Machine
+ Kernel과 C값을 바꾸어가며 적절한 모델 찾기
+ C : 오류를 얼마나 허용할 것인지 (C ↓ : 소프트 마진) 

```python
#Support Vecotr Machine - linear kernel
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

model = SVC(kernel = 'linear', C = 1)
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", round(accuracy,4))
```

```python
#Support Vecotr Machine - radial basis function
model = SVC(kernel = 'rbf', C = 10)
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", round(accuracy,4))
```

### Naïve bayes classifier
+ Categorical Naïve bayes 모델과 Gaussian Naïve bayes 모델에 적용
+ CategoricalNB을 사용하기 위하여 Categorical 하지 못한 피처를 제거(absences)

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", round(accuracy,4))
```

```python
# 피처들을 모두 Categorical 하게 만들어주기
# Categorical하지 못한 특성을 갖는 피처 제거
my_train = X_train.drop(['absences'], axis = 1)
my_test = X_test.drop(['absences'], axis = 1)
```

```python
from sklearn.naive_bayes import CategoricalNB
model = CategoricalNB()
model.fit(my_train, Y_train)
y_pred = model.predict(my_test)

accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", round(accuracy,4))
```

### Decision tree for classification

### Subset Selection - Forward Stepwise Selection
+ accuracy를 높이기 위해 Foward Stepwise Selection을 수행
+ Forward Stepwise Selection : 피처를 하나씩 추가해가며 loss(RSS)를 비교해 가장 작은 loss를 주는 피처를 하니씩 추가
+ Forward Stepwise Selection의 방법을 활용해 accuray를 가장 높게 하는 피처들을 하나씩 추가

```python
cols = ['school','sex','age','address','famsize', 'Mjob','Fjob','reason','traveltime','studytime','failures',
        'romantic','famrel', 'freetime','goout','Dalc','Walc','health','absences','G1','G2','G3']

model = SVC(kernel = 'rbf', C = 10)
selected_feature = []
data_x = ['label']

for i in range(0,22):
    acc = 0
    for j in range(0,len(cols)):    
        new_acc = 0
        data_x.append(cols[j])
        my_data = data[data_x]

        y = my_data['label']
        x = my_data.drop(['label'], axis = 1)
        
        # data를 8:2로 train data와 test data로 분리
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 39)

        #모델 훈련, 예측
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
        new_acc = accuracy_score(y_test, y_pred)
            
        if(acc <= new_acc):
            acc = new_acc
            current_feature = j
            # print(feature, new_acc)
        data_x.pop()

    data_x.append(cols[current_feature])
    selected_feature.append(cols[current_feature])
    
    print("# selected feature : " , len(selected_feature))
    print(selected_feature)
    print("Accuracy: ", round(acc,4) ,'\n')
    cols.pop(current_feature)
```
